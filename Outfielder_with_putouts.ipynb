## The Question:
Which outfeild position has the most put outs? 
Is it becuase they get more chances or the better players make the plays?

## The plan:
The plan will be to gather all outs that have an outfielder listed in the description (ie. "Cody Bellinger flies out to left **fielder** Marwin Gonzalez.").

Here are the types of outs that we are looking for in the description of the at bat:
Groundout, Flyout, Pop Out, Lineout, Forceout, Grounded Into DP, Bunt Pop Out, Bunt Groundout, Fielders Choice Out, Sac Fly DP, Sacrifice Bunt DP, and Bunt Lineout 

We'll weed out any events that don't include the date, because I think its important to look at the time frame of the events as well. 

```python
import pandas as pd
import numpy as np

all_outs = pd.read_csv("data/outfield-putouts.csv")
all_outs.head()
```

Above is an example of all the data that we have.

```python
trimed_down = all_outs[["event", "atbat_des", "start_tfs_zulu"]].copy()
trimed_down.head()
```

Here we have the data that we need, but let's just grab who the out goes to.

```python
trimed_down.count()
```

Looks like we have almost 720,000 outs created from the outfield.

```python
center_fielder = 0
right_fielder = 0
left_fielder = 0

for index, row in trimed_down.iterrows():
    if "center fielder" in row['atbat_des']:
        center_fielder += 1
    if "right fielder" in row['atbat_des']:
        right_fielder += 1
    if "left fielder" in row['atbat_des']:
        left_fielder += 1
        
print("Center Fielder outs: ", center_fielder)
print("Right Fielder outs: ", right_fielder)
print("Left Fielder outs: ", left_fielder)
```
